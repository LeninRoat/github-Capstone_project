```python
#send request
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
page = requests.get(url)
soup = BeautifulSoup(page.text,'html.parser')
print(soup)
```

```python
# finds table name in wiki and creates a variable 
table = soup.find('table',{'class':'wikitable'})
print(table)                                                
```

```python
#pandas reads the variable created above
df = pd.read_html(str(table))
df
```

```python
#creates the columns and rows for dataframe
rows = [v.text.replace('\n','') for v in table.find_all('tr')]
column_names = [v.text.replace('\n','') for v in table.find_all('th')]
neighbourhoods = pd.DataFrame(columns=column_names)
neighbourhoods
```

```python
#appends the dataframe with corresponding information
for tr_d in table.find_all('tr'):
        row_data=[]
        for td_d in tr_d.find_all('td'):
            row_data.append(td_d.text.strip())
        if len(row_data)==3:
            neighbourhoods.loc[len(neighbourhoods)] = row_data
neighbourhoods.head()           
```

```python
#filters out Not assigned
df = neighbourhoods[neighbourhoods['Borough'] != 'Not assigned']
df
```

```python
#creates temp dataframe with a new column that merges bourgh and neighbourhood on postal code 
df2=df.groupby('Postal Code')['Neighbourhood'].apply(lambda x: "%s" % ', '.join(x))
df2=df2.reset_index(drop=False)
df2.rename(columns={'Neighbourhood':'Neighbourhood_joined'},inplace=True)
df2
```

```python
#creates new dataframe by joining the temp table and the original table and uses the new column created in the temp dataframe and changes the name to the original one
data_join = pd.merge(df, df2, on='Postal Code')
data_join.drop(['Neighbourhood'],axis=1,inplace=True)
data_join.drop_duplicates(inplace=True)
data_join.rename(columns={'Neighbourhood_joined':'Neighbourhood'},inplace=True)
data_join.head()
```
